## Question 2. Which city has the cleanest air ?
### Tip: consider adapting the following example: https://seaborn.pydata.org/examples/grouped_boxplot.html

```python
fig, ax = plt.subplots(1, figsize=(12,6))

ax = sns.boxplot(
    x='city',
    y='value',
    data=airQualityCities,
    fliersize=0,
    palette='deep',
    ax=ax)

ax.set_ylim([0, 750])
ax.set_ylabel("City", fontsize=18)
ax.set_xlabel("value")

sns.despine(offset=10)


plt.show()
```

## Question 3. Which location is the most poluted in London ?

```python
airQualityCities[airQualityCities['city']=='London']
```

```python
londonpolution=airQualityCities[airQualityCities['city']=='London']
```

```python
londonpolution
```

```python
fig, ax = plt.subplots(1, figsize=(14,7))

ax = sns.boxplot(
    x='location',
    y='value',
    data=londonpolution,
    fliersize=0,
    palette='deep',
    ax=ax)

ax.set_ylim([0, 750])
ax.set_ylabel("value", fontsize=18)
ax.set_xlabel("location")

sns.despine(offset=10)


plt.show()
```

## Question 4. Plot the evolution of air polution accross time in locations in London
### Tip: Consider using Relplot: https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
londonpolution
```

```python
airqualityevolution=londonpolution.groupby(['location','date.local'])['value'].mean()
airqualityevolution.unstack(level=0).plot(kind='line',subplots=False, figsize=(15,15))
```
